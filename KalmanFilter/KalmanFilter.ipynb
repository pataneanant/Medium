```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Signal parameters
t = np.linspace(0, 1, 1000)  # Time vector, 1 second duration sampled at 1000 Hz
frequency = 2  # Frequency of the signal in Hz
amplitude = 1  # Amplitude of the signal

# Generate a clean accelerometer signal (sine wave in this case)
clean_signal = amplitude * np.sin(2 * np.pi * frequency * t)

# Additive noise
additive_noise = np.random.normal(0, 0.2, len(t))
noisy_signal_additive = clean_signal + additive_noise

# Multiplicative noise
multiplicative_noise = clean_signal * np.random.normal(0, 0.2, len(t))
noisy_signal_multiplicative = clean_signal + multiplicative_noise

# Impulse noise
impulse_noise = np.copy(clean_signal)
indices = np.random.choice(len(t), size=20, replace=False)
impulse_noise[indices] += np.random.normal(0, 2, 20)
noisy_signal_impulse = impulse_noise

# Quantization noise
quantization_levels = 32
quantization_noise = np.random.uniform(-0.5, 0.5, len(t)) / quantization_levels
noisy_signal_quantization = clean_signal + quantization_noise

# Plotting

# Plot 1: Clean Accelerometer Signal
plt.figure(figsize=(8, 4))
plt.plot(t, clean_signal, label='Clean Signal', color='blue', linewidth=2)
plt.title('Clean Accelerometer Signal', fontsize=16)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.legend(loc='upper right')
plt.show()

# Plot 2: Accelerometer Signal with Additive Noise
plt.figure(figsize=(8, 4))
plt.plot(t, noisy_signal_additive, label='Noisy Signal (Additive)', color='red', linewidth=2)
plt.title('Accelerometer Signal with Additive Noise', fontsize=16)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.legend(loc='upper right')
plt.show()

# Plot 3: Accelerometer Signal with Multiplicative Noise
plt.figure(figsize=(8, 4))
plt.plot(t, noisy_signal_multiplicative, label='Noisy Signal (Multiplicative)', color='green', linewidth=2)
plt.title('Accelerometer Signal with Multiplicative Noise', fontsize=16)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.legend(loc='upper right')
plt.show()

# Plot 4: Accelerometer Signal with Impulse Noise
plt.figure(figsize=(8, 4))
plt.plot(t, noisy_signal_impulse, label='Noisy Signal (Impulse)', color='purple', linewidth=2)
plt.title('Accelerometer Signal with Impulse Noise', fontsize=16)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.legend(loc='upper right')
plt.show()

# Plot 5: Accelerometer Signal with Quantization Noise
plt.figure(figsize=(8, 4))
plt.plot(t, noisy_signal_quantization, label='Noisy Signal (Quantization)', color='orange', linewidth=2)
plt.title('Accelerometer Signal with Quantization Noise', fontsize=16)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.legend(loc='upper right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from filterpy.kalman import KalmanFilter

measurements = data['x'][:20500]  \

# Create a Kalman filter with 1D state (position) and 1D measurement (acceleration)
kf = KalmanFilter(dim_x=1, dim_z=1)

# Define the state transition matrix
kf.F = np.array([[1]])

# Define the measurement matrix
kf.H = np.array([[1]])

# Set the initial state and covariance
kf.x = measurements[0]
kf.P *= 1e3

# Set the process noise covariance
kf.Q = np.array([[1e-3]])

# Set the measurement noise covariance
kf.R = np.array([[0.1]])

# Create empty lists to store the filtered and predicted values
filtered_state_means = []
predicted_state_means = []

# Apply the Kalman filter to each measurement
for measurement in measurements:
    kf.predict()
    kf.update(measurement)
    filtered_state_means.append(kf.x[0])
    predicted_state_means.append(kf.x[0])

# Plot the original and filtered signals
plt.figure(figsize=(10, 6))
plt.plot(measurements, label='Original Signal', color='orange', linewidth=2)
plt.plot(filtered_state_means, label='Filtered Signal', color='blue', linewidth=2)
plt.title('Original vs. Filtered Signal')
plt.xlabel('Index')
plt.ylabel('Acceleration (x)')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

original_signal = data['x'][:20500]
filtered_signal = filtered_state_means

# Calculate Signal-to-Noise Ratio (SNR)
snr_original = np.mean(original_signal) / np.std(original_signal)
snr_filtered = np.mean(filtered_signal) / np.std(filtered_signal)
improvement_snr = snr_filtered - snr_original

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(original_signal, filtered_signal)

# Print results
print(f"Original SNR: {snr_original}")
print(f"Filtered SNR: {snr_filtered}")
print(f"Improvement in SNR: {improvement_snr}")
print(f"Mean Squared Error (MSE): {mse}")

```
